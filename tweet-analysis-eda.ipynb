**Data Cleaning Process**


1.Remove Unamed column as we have another index

2.Typecast DataType of date of Date type

3.Remove Columns language,link as they are no help.Also same language for entire tweets.

4.Remove NaN rows as we cannot do anything about Usernames 

5.Segregate date into month,year,date and add new colums to dataframe 

6.Change Datatype of replies_count,likes_count,retweets_count


```python
df.drop(['Unnamed: 0','language','link'],axis=1)
```

```python
df.dropna(subset=['username'],inplace=True)
```

```python
df['date'] = pd.to_datetime(df['date'])
df.date.dtype
```

```python
df['year'] = df['date'].dt.year.fillna(0)
df['month'] = df['date'].dt.month.fillna(0)
df['day'] = df['date'].dt.day.fillna(0)
```

```python
df['day Name'] = df['date'].dt.day_name()
df.head()
```

```python
df['phone name'] = df['phone name'].str.replace(',', '')
df['username'] = df['username'].str.replace(',', '')
df['username'] = df['username'].str.strip()
df['phone name'] = df['phone name'].str.strip()
```

```python
df=df.astype({'replies_count':'int','retweets_count':'int','likes_count':'int'})
```

***Basic_EDA Analysis***

1.Top 10 phones with hghest no of tweets

2.Top 10 phones with highest no of likes,reweets,replies

3.Top 10 usernames with highest interaction

4.Highest number of tweets in a year,day,date

5.Highest no of tweets on a particular date

6.Find top 5 users with most tweets

7.Find most frequent word in tweets

```python
sns.countplot(data=df,x='phone name',order=df['phone name'].value_counts().index[:10])
plt.xticks(rotation='75')
plt.show()
```

```python
df1=df.groupby('phone name')['likes_count'].sum().to_frame()
df1 = df1.sort_values('likes_count',ascending=False)
plt.bar(df1.index[:10],df1.likes_count[:10])
plt.xticks(rotation='90')
plt.show()
```

```python
df1=df.groupby('phone name')['replies_count'].sum().to_frame()
df1 = df1.sort_values('replies_count',ascending=False)
plt.bar(df1.index[:10],df1.replies_count[:10])
plt.xticks(rotation='90')
plt.show()
```

```python
df1=df.groupby('phone name')['retweets_count'].sum().to_frame()
df1 = df1.sort_values('retweets_count',ascending=False)
plt.bar(df1.index[:10],df1.retweets_count[:10])
plt.xticks(rotation='90')
plt.show()
```
