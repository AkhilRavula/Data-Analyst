**Top 5 clubs from where most players come?**

```python
sns.countplot(data=df,x='Club',order =df.Club.value_counts().index[:5] ,palette='Set3')
plt.xticks(rotation='90')
plt.show()
```

```python
df[df['Club']=='Manchester United']['Position'].value_counts()
```

**Top 5 clubs with their positio status**

```python
new_df=df.set_index('Club')
TTlist=df.Club.unique().tolist()
TTlist = [i for i in TTlist if i not in ['Paris Saint-Germain','Manchester United','Manchester City','Chelsea FC'
                                         ,'Tottenham Hotspur']]

new_df=new_df.drop(TTlist)
new_df=new_df.reset_index('Club')


plt.figure(figsize=(12,8))
sns.countplot(data=new_df,x='Position',order =df.Position.value_counts().index,hue='Club',palette='Set2')
plt.xticks(rotation='90')
plt.legend(loc='upper right')
plt.show()
```

**as datatype is int we can add two columns without cast**

```python
df["Any Card"] = df["Yellow Cards"] + df["Second Yellow Cards"] + df["Red Cards"]
df.head(8)
```

**Top 5 players with most no .of card**

```python
sort_v = df.sort_values('Any Card',ascending=False)
sort_v['Name'][:4]
```

```python
df[df['Club']=='AC Milan']['Any Card'].sum()
```

**Top 5 clubs with most no of cards**

```python
g=df.groupby('Club',sort='Any Card')
g.get_group('Manchester United')
g.sum().sort_values('Any Card',ascending=False)['Any Card'][:5]
```

```python
data=g.sum().sort_values('Any Card',ascending=False)['Any Card'][:5]
lt=data.tolist()
lt1=data.index.tolist()
plt.bar(lt1,lt,width=0.5,color='green')
plt.xticks(rotation='90')
plt.show()
```
