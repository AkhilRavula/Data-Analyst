As no data cleanup is required we shall proceed to EDA.
**No.of Winning candidates from Diff Caste's**

```python
sns.countplot(data=df,x='Type',order = df.Type.value_counts().index,palette='Set1')
plt.show()
```

**Top 5 candidates with highest/lowest Poll%**

```python
sort_poll = df.sort_values('Poll%',ascending=False)
sort_poll
```

```python

sort_v=sort_poll[['Poll%','AC Name']][:5]
sort_v
```

```python
sort_poll = df.sort_values('Poll%',ascending=True)
sort_h=sort_poll[['Poll%','AC Name']][:5]
sort_h
```

**No. of winning candidates from a single party with diff caste's**

```python
sns.set_theme(style="darkgrid")
sns.countplot(data=df,x='Party',hue='Type',order=df.Party.value_counts().index[:7])
plt.xticks(rotation='90')
plt.legend(loc='upper right')
plt.show()
```

**Top 5 highest/lowest wins by Margin**

```python
sort_m=df.sort_values('Margin',ascending=False)
sort_m
```

```python
sort_m[['Winning Candidate','Margin']][:5]
```

```python
sort_m=df.sort_values('Margin',ascending=True)
sort_m[['Winning Candidate','Margin']][:5]
```

**District having highest AC's**

```python
g=df.groupby('District')
g.count().sort_values('AC Name',ascending=False)['AC No.'][:5]
```
