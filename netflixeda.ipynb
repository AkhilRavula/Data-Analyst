```python
df['date_added'] = pd.to_datetime(df['date_added'])
type(df['date_added'][0])
```

```python
df['date_added'][0]
#date datatype achieved
```

```python
#dropping all rows with NAN in date_added column
df.dropna(subset=['date_added'],inplace=True)
```

```python
df.isna().sum()
#missing values is zero in date_added
```

```python
df.director.fillna('missing',inplace=True)
```

```python
df.cast.fillna('missing',inplace=True)
df.country.fillna('missing', inplace = True)
df.rating.fillna('missing', inplace = True)
```

```python
cast_h = []

for i in range(len(df)):
    cast_h.extend(df.cast.iloc[i].split(','))
#print(cast_h)
    
```

```python
cast_direc = {}
for i in cast_h:
    if i in cast_direc:
        cast_direc[i] += 1
    else:
        cast_direc[i] = 1
```

```python
cast_direc = sorted(cast_direc.items(), key = lambda item: item[1], reverse = True)
```

```python
cast_direc = dict(cast_direc)
listx = list(cast_direc.keys())     
listy = list(cast_direc.values())

print(len(cast_direc))
```

```python
df['added_month'] = df['date_added'].dt.month.fillna(0)
df['added_date'] = df['date_added'].dt.day.fillna(0)
df['added_year'] = df['date_added'].dt.year.fillna(0)
df.columns
```

```python
listed_h = []

for i in range(len(df)):
    listed_h.extend(df.listed_in.iloc[i].split(','))
```

```python
listed = {}
for i in listed_h:
    if i in listed:
        listed[i] += 1
    else:
        listed[i] = 1
```
