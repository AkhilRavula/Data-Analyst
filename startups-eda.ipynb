```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('/kaggle/input/unicorn-startups/Startups in 2021 end.csv')
df.head()
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
def nans(df): return df[df.isnull().any(axis=1)]
nans(df)
```

```python
#df.City.fillna('missing',inplace=True)
df['City'].fillna(df['Country'], inplace=True)
df['Select Investors'].fillna('missing',inplace=True)
```

```python
df=df.drop('Unnamed: 0',axis=1)
```

```python
df['Date Joined']=pd.to_datetime(df['Date Joined'])
```

```python
df['Year Joined']=df['Date Joined'].dt.year.fillna(0)
```

```python
nans(df)
```

```python
df.columns
```

```python
df['Country'] = df['Country'].str.replace(',', '')
df['City'] = df['City'].str.replace(',', '')
df['Country'] = df['Country'].str.strip()
df['City'] = df['City'].str.strip()
```

```python
investors = []

for i in range(len(df)):
    investors.extend(df['Select Investors'].iloc[i].split(','))
```

```python
dict_inves = {}
for i in investors:
    i=str(i).strip()
    if i in dict_inves:
        dict_inves[i] += 1
    else:
        dict_inves[i] = 1
```

```python
dict_inves = sorted(dict_inves.items(), key = lambda item: item[1], reverse = True)
```

```python
dict_inves=dict(dict_inves)
list1=list(dict_inves.values())
list2=list(dict_inves.keys())
```

```python
plt.bar(list2[:15],list1[:15],color='g')
plt.xticks(rotation='90')
plt.ylabel('No. of startups funded')
plt.show()
```
